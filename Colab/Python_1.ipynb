## Алгоритмы

Как найти приложение на телефоне?

1. Взяли телефон в руку, включили экран
2. Смотрим на 1ую иконку на экране. Проверяем наше ли приложение.
3. Смотрим на 2ую иконку...
4. ...
5. Смотрим на nую иконку на экране.
6. Если мы не нашли, то переходим на следующий экран. Повторяем с пункта 2.
7. Если мы прошли все экраны, то прекртатить искать (( Мы не нашли приложение.

## Основы Python

```
class Телефон:
  def __init__(self):
    self.screen=0
    self.apps = [
      ["Google Calendaar", "Yandex Disk", "Zotero", "Bolt", "Zoom"],
      ["Tinder", "Discord", "Instagram", "Telegram"],
      ["Google Drive", "TikTok", "Zotero"]
    ]
  def какой_экран(self):
    print(f"ты на экране номер {self.screen}")

  def дай_мне_номер_экрана(self):
    return self.screen

  def перейди_на_следующий_экран(self):
    self.screen = self.screen+1%3

  def дай_мне_количчество_экранов(self):
    return len(self.apps)

  def дай_мне_название_приложения_на_этом_месте(self, position):
    if position < len(self.apps[self.screen]):
      return self.apps[self.screen][position]

  def дай_мне_количество_приложений_на_этом_экране(self, position):
    return len(self.apps[self.screen])

  def дай_мне_количество_приложений_на_этом_экране(self):
    return len(self.apps[self.screen])

  def проверь_на_этом_месте_находится_приложение(self, position, app):
    if position < len(self.apps[self.screen]):
      return self.apps[self.screen][position] == app
    return False
```

### Управление (control flow)

```
# Условия
condition = True
if condition:
  # если условие True
  pass
else:
  # если условие False
  pass

# pass - специальное слово, чтобы заполнить пустые блоки
# >, <, >=, <=, ==, !=
# and, or, not

# x and y -> z (и)
# False False -> False
# False True  -> False
# True  False -> False
# True  True  -> True

# x or y -> z (или)
# False False -> False
# False True  -> True
# True  False -> True
# True  True  -> True

# not x -> z (не)
# False -> True
# True  -> False
```

### Императивное программировнаие

```
# Pythonic стиль и императивное программировнаие

l = [5,6,7,8]
total = 0
for i in range(len(l)):
  val = l[i]
  total += val

print(total)

# Чуть проще
total = 0
for x in [5,6,7,8]:
  total += x

print(total)

# Pythonic
total = sum([5,6,7,8])
print(total)
```
